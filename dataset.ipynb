```python
import cv2
import shutil
import random
from pathlib import Path
```

```python
# Directorios del conjunto de datos original
ROOT = Path("/mnt/d/TFM/German dataset/FullIJCNN2013")
IMAGES_DIR = ROOT
LABELS_SRC = ROOT / "labels_yolo"

OUT = Path("/mnt/d/TFM/dataset_YOLO/")
```

```python
# Crear directorios de salida
for sub in ["train", "val", "test"]:
    (OUT / "images" / sub).mkdir(parents=True, exist_ok=True)
    (OUT / "labels" / sub).mkdir(parents=True, exist_ok=True)
```

```python
# Separación de imágenes en conjuntos aleatorios de 70%train/20%val/10%test
random.seed(42)

all_images = sorted([p for p in IMAGES_DIR.glob("*.ppm")])
random.shuffle(all_images)

n = len(all_images)
n_train = int(0.7 * n)
n_val = int(0.2 * n)
train_imgs = all_images[:n_train]
val_imgs = all_images[n_train:n_train + n_val]
test_imgs = all_images[n_train + n_val:]



splits = {
    "train": train_imgs,
    "val": val_imgs,
    "test": test_imgs,
}

for split_name, img_list in splits.items():
    for img_path in img_list:
        base = img_path.stem
        label_src = LABELS_SRC / f"{base}.txt"

        # copiar imagen
        shutil.copy2(img_path, OUT / "images" / split_name / img_path.name)

        # copiar etiqueta (si existe)
        if label_src.exists():
            shutil.copy2(label_src, OUT / "labels" / split_name / label_src.name)
        else:
            print(f"[AVISO] No hay etiqueta para {img_path.name}")


print("Split train/val/test creado en", OUT)
```
